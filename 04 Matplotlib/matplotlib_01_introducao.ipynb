## plot()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y)
plt.show()
```

### Legendas

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.xlabel("Nome do Eixo X")
plt.ylabel("Nome do Eixo Y")
plt.show()
```

### Escala

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.xlabel("Nome do Eixo X")
plt.ylabel("Nome do Eixo Y")
plt.xticks([1,3,5])
plt.yticks([-30,0,30,60,90,120,150,180,200])
plt.show()
```

### Propriedades de Linha

https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas", linestyle="dashed", color="m", linewidth="3") # Nome do tipo da linha ou sinbolos: --
plt.legend()
plt.xlabel("Nome do Eixo X")
plt.ylabel("Nome do Eixo Y")
plt.xticks([1,3,5])
plt.yticks([-30,0,30,60,90,120,150,180,200])
plt.show()
```

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas", linestyle="dashed", color="m", linewidth="3") # Nome do tipo da linha ou sinbolos: --
plt.legend()
plt.xlabel("Nome do Eixo X")
plt.ylabel("Nome do Eixo Y")
plt.xticks([1,3,5])
plt.yticks([-30,0,30,60,90,120,150,180,200])
plt.title("Seu animal!", color="r")
plt.show()
```

## Tipos de Gráficos 

### scatter - gráfico de dispersão

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
plt.scatter(x,y)
plt.show()
```

### bar - gráfico de barras

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
plt.bar(x,y)
plt.show()
```

### pie - gráfico de pizza

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
z = [23,5,100,2]
plt.pie(z)
plt.show()
```

## Mudando o tipo de marcador do gráfico

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x,y, marker="v")
```

## Mudando os limites do gráfico

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html

```python
plt.plot(x,y)
plt.axis(xmin=-1,xmax=8,ymin=-50,ymax=300)
plt.show()
```

```python
plt.plot(x,y)
plt.axis("square")
plt.show()
```

## Na prática - dataset Pokemon e UFO

```python
import pandas as pd

pokemon = pd.read_csv('Pokemon.csv')
pokemon
```

```python
pokemon_attack = pokemon[['Name', 'Attack']]
pokemon_attack
```

```python
x = pokemon_attack['Attack']
y = pokemon_attack['Name']
plt.bar(x,y)
plt.show
```

```python
pokemon_attack.sort_values('Attack', ascending=False).head(6)
```

```python
pokemon_maiores = pokemon_attack.sort_values('Attack', ascending=False).head(6)
x = pokemon_maiores['Name']
y = pokemon_maiores['Attack']
plt.title("Maiores Pokemons")
plt.bar(x,y)
plt.show()
```

```python
pokemon_maiores = pokemon_attack.sort_values('Attack', ascending=False).head(6)
x = pokemon_maiores['Name']
y = pokemon_maiores['Attack']
plt.figure(figsize=(15,5))
plt.bar(x,y)
plt.show()
```

```python
pokemon_maiores = pokemon_attack.sort_values('Attack', ascending=False).head(6)
x = pokemon_maiores['Name']
y = pokemon_maiores['Attack']
plt.xticks(rotation=50, ha='right')
plt.bar(x,y)
plt.show()
```

```python
y_attack = pokemon['Attack'].head(6)
y_defense = pokemon['Defense'].head(6)
x = pokemon['Name'].head(6)
plt.figure(figsize=(15,5)) #Largura
plt.plot(x,y_attack, label='Attack')
plt.plot(x,y_defense, label='Defense')
plt.legend()
plt.show()
```

```python
y_attack = pokemon['Attack']
y_defense = pokemon['Defense']
x = pokemon['Name']
plt.figure(figsize=(15,5)) #Largura
plt.scatter(x,y_attack, label='Attack')
plt.scatter(x,y_defense, label='Defense')
plt.legend()
plt.show()
```
