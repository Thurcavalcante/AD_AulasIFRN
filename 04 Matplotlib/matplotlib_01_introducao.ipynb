# Matplotlib

- Biblioteca popular para visualização e geração de gráficos, histogramas e outras visualizações profissionais;
- Possui recursos interativos como zoom e visão panorâmica, além de possibilitar exportar as imagens para vários formatos como vetor comum, pdf, svg, jpg, png, bmp, gif, entre outros.

Site Oficial: https://matplotlib.org/

Documentação: https://matplotlib.org/stable/index.html

Passos iniciais: https://matplotlib.org/stable/users/explain/quick_start.html

Instalação: pip install matplotlib

```python
import matplotlib as mpl
```

```python
import matplotlib.pyplot as plt
```

```python
'''
Se o gráfico abrir uma nova janela
'''
%matplotlib inline
```

## plot()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y)
plt.show()
```

### Legendas

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.xlabel("Nome do Eixo X")
plt.ylabel("Nome do Eixo Y")
plt.show()
```

### Escala

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.xlabel("Nome do Eixo X")
plt.ylabel("Nome do Eixo Y")
plt.xticks([1,3,5])
plt.yticks([-30,0,30,60,90,120,150,180,200])
plt.show()
```

### Propriedades de Linha

https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas", linestyle="dashed", color="m", linewidth="3") # Nome do tipo da linha ou sinbolos: --
plt.legend()
plt.xlabel("Nome do Eixo X")
plt.ylabel("Nome do Eixo Y")
plt.xticks([1,3,5])
plt.yticks([-30,0,30,60,90,120,150,180,200])
plt.show()
```

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas", linestyle="dashed", color="m", linewidth="3") # Nome do tipo da linha ou sinbolos: --
plt.legend()
plt.xlabel("Nome do Eixo X")
plt.ylabel("Nome do Eixo Y")
plt.xticks([1,3,5])
plt.yticks([-30,0,30,60,90,120,150,180,200])
plt.title("Seu animal!", color="r")
plt.show()
```

## Tipos de Gráficos 

### scatter - gráfico de dispersão

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
plt.scatter(x,y)
plt.show()
```

### bar - gráfico de barras

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
plt.bar(x,y)
plt.show()
```

### pie - gráfico de pizza

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
z = [23,5,100,2]
plt.pie(z)
plt.show()
```

## Mudando o tipo de marcador do gráfico

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x,y, marker="v")
```

## Mudando os limites do gráfico

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html

```python
plt.plot(x,y)
plt.axis(xmin=-1,xmax=8,ymin=-50,ymax=300)
plt.show()
```

```python
plt.plot(x,y)
plt.axis("square")
plt.show()
```
